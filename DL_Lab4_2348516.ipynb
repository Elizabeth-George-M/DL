```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/dl5.csv")
```

```python
data.columns
```

```python
data = data.drop(columns=['State'])
```

```python
from sklearn.preprocessing import LabelEncoder

columns=['Voice mail plan','International plan','Churn']

for i in columns:
  encoder = LabelEncoder()
  data[i] = encoder.fit_transform(data[i])
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

dataScale = data.drop(columns=['Churn'])
for i in dataScale.columns:
    scale = StandardScaler()
    data[i] = scale.fit_transform(data[i].values.reshape(-1, 1))
```

```python
data.head()
```

```python
x = data.drop(columns=['Churn'])
y = data['Churn']
```

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt

epochs = 10


X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)

# Baseline DNN model without dropout
model_baseline = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Model with Dropout
model_with_dropout = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Model with Layer-wise Dropout
model_layerwise_dropout = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

# Model with Monte Carlo Dropout
model_monte_carlo_dropout = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

# Compile models
models = [model_baseline, model_with_dropout, model_layerwise_dropout, model_monte_carlo_dropout]

for model in models:
    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Train models
history = []
for model in models:
    hist = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_val, y_val), verbose=0)
    history.append(hist.history)

# Evaluate models
metrics_dict = {}
for i, model in enumerate(models):
    y_pred = model.predict(X_val)
    y_pred_binary = (y_pred > 0.5).astype(int)

    accuracy = accuracy_score(y_val, y_pred_binary)
    f1 = f1_score(y_val, y_pred_binary)
    roc_auc = roc_auc_score(y_val, y_pred)

    metrics_dict[f"Model {i+1}"] = {'Accuracy': accuracy, 'F1 Score': f1, 'ROC AUC': roc_auc}

# Comparison and Visualization
for i, model in enumerate(models):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history[i]['accuracy'], label='Training Accuracy')
    plt.plot(history[i]['val_accuracy'], label='Validation Accuracy')
    plt.title(f'Model {i+1} - Accuracy Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history[i]['loss'], label='Training Loss')
    plt.plot(history[i]['val_loss'], label='Validation Loss')
    plt.title(f'Model {i+1} - Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Print performance metrics
for model, metrics in metrics_dict.items():
    print(f"\n{model} Metrics:")
    for metric, value in metrics.items():
        print(f"{metric}: {value}")
```
